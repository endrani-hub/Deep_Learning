```python
weight = 0.5
ip = 0.5
goal_pred = 0.8
step_size = 0.001
```

```python
def neural_net(ip , weight):
    pred = ip * weight
    return pred
```

```python
for i in range(1101):
    
    pred = neural_net(ip , weight)
    err = (goal_pred - pred) ** 2
    
    pred_up = neural_net(ip , weight + step_size)
    err_up = (goal_pred - pred_up) ** 2
    
    pred_down = neural_net(ip , weight - step_size)
    err_down = (goal_pred - pred_down) ** 2
    
    if(err_up < err_down):
        weight = weight + step_size
        print("Error is" , err_up,"\tPrediction is" , pred_up)
    elif(err_down < err_up):
        weight = weight - step_size
        print("Error is" , err_down , "\tPrediction is" , pred_down)
```
